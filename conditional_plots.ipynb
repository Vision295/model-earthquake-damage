```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def load_and_merge_data(train_values_path, train_labels_path):
    """Load training data and labels, merge them"""
    df_values = pd.read_csv(train_values_path)
    df_labels = pd.read_csv(train_labels_path)
    
    # Merge on building_id
    df = df_values.merge(df_labels, on='building_id', how='inner')
    
    return df

def plot_conditional_probabilities_bool(feature_col, prediction_col, feature_name):
    """
    Plot P(feature | prediction), P(prediction | feature), and P(feature)
    for boolean features
    """
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    fig.suptitle(f'Conditional Probabilities: {feature_name}', fontsize=16, fontweight='bold')
    
    # Get unique prediction values
    pred_vals = sorted(prediction_col.unique())
    
    # Plot 1: P(feature | prediction) - proportion of True for each damage grade
    ax1 = axes[0]
    probs = []
    for pred in pred_vals:
        mask = prediction_col == pred
        prob_true = feature_col[mask].mean()  # Mean of boolean = proportion of True
        probs.append(prob_true)
    
    ax1.bar(range(len(pred_vals)), probs, color='steelblue', alpha=0.7)
    ax1.set_xticks(range(len(pred_vals)))
    ax1.set_xticklabels([f'Grade {p}' for p in pred_vals])
    ax1.set_ylabel('P(Feature = True | Prediction)')
    ax1.set_title('P(Feature | Prediction)')
    ax1.set_ylim([0, 1])
    ax1.grid(axis='y', alpha=0.3)
    
    # Plot 2: P(prediction | feature) - distribution of damage grade for True vs False
    ax2 = axes[1]
    width = 0.35
    x = np.arange(len(pred_vals))
    
    # For feature = True
    true_dist = []
    for pred in pred_vals:
        count = ((feature_col == True) & (prediction_col == pred)).sum()
        total = (feature_col == True).sum()
        true_dist.append(count / total if total > 0 else 0)
    
    # For feature = False
    false_dist = []
    for pred in pred_vals:
        count = ((feature_col == False) & (prediction_col == pred)).sum()
        total = (feature_col == False).sum()
        false_dist.append(count / total if total > 0 else 0)
    
    ax2.bar(x - width/2, true_dist, width, label='Feature = True', alpha=0.7)
    ax2.bar(x + width/2, false_dist, width, label='Feature = False', alpha=0.7)
    ax2.set_xticks(x)
    ax2.set_xticklabels([f'Grade {p}' for p in pred_vals])
    ax2.set_ylabel('P(Prediction | Feature)')
    ax2.set_title('P(Prediction | Feature)')
    ax2.legend()
    ax2.grid(axis='y', alpha=0.3)
    
    # Plot 3: P(feature) - marginal distribution
    ax3 = axes[2]
    counts = feature_col.value_counts().sort_index()
    probs_marginal = counts / counts.sum()
    ax3.bar([0, 1], probs_marginal.values, color='coral', alpha=0.7)
    ax3.set_xticks([0, 1])
    ax3.set_xticklabels(['False', 'True'])
    ax3.set_ylabel('P(Feature)')
    ax3.set_title('P(Feature) - Marginal')
    ax3.set_ylim([0, 1])
    ax3.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_conditional_probabilities_categorical(feature_col, prediction_col, feature_name):
    """
    Plot conditional probabilities for categorical features
    """
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'Conditional Probabilities: {feature_name}', fontsize=16, fontweight='bold')
    
    pred_vals = sorted(prediction_col.unique())
    feature_vals = sorted(feature_col.dropna().unique(), key=str)
    
    # Plot 1: P(feature | prediction) - heatmap
    ax1 = axes[0]
    cond_probs = []
    for pred in pred_vals:
        mask = prediction_col == pred
        dist = feature_col[mask].value_counts(normalize=True)
        probs = [dist.get(fv, 0) for fv in feature_vals]
        cond_probs.append(probs)
    
    cond_probs = np.array(cond_probs).T
    im1 = ax1.imshow(cond_probs, aspect='auto', cmap='YlOrRd')
    ax1.set_yticks(range(len(feature_vals)))
    ax1.set_yticklabels(feature_vals)
    ax1.set_xticks(range(len(pred_vals)))
    ax1.set_xticklabels([f'Grade {p}' for p in pred_vals])
    ax1.set_ylabel('Feature Value')
    ax1.set_xlabel('Prediction')
    ax1.set_title('P(Feature | Prediction)')
    plt.colorbar(im1, ax=ax1)
    
    # Plot 2: P(prediction | feature) - heatmap
    ax2 = axes[1]
    cond_probs2 = []
    for fv in feature_vals:
        mask = feature_col == fv
        dist = prediction_col[mask].value_counts(normalize=True)
        probs = [dist.get(pv, 0) for pv in pred_vals]
        cond_probs2.append(probs)
    
    cond_probs2 = np.array(cond_probs2)
    im2 = ax2.imshow(cond_probs2, aspect='auto', cmap='YlGnBu')
    ax2.set_yticks(range(len(feature_vals)))
    ax2.set_yticklabels(feature_vals)
    ax2.set_xticks(range(len(pred_vals)))
    ax2.set_xticklabels([f'Grade {p}' for p in pred_vals])
    ax2.set_ylabel('Feature Value')
    ax2.set_xlabel('Prediction')
    ax2.set_title('P(Prediction | Feature)')
    plt.colorbar(im2, ax=ax2)
    
    # Plot 3: P(feature) - marginal distribution
    ax3 = axes[2]
    marginal = feature_col.value_counts(normalize=True).reindex(feature_vals, fill_value=0)
    ax3.bar(range(len(feature_vals)), marginal.values, color='seagreen', alpha=0.7)
    ax3.set_xticks(range(len(feature_vals)))
    ax3.set_xticklabels(feature_vals, rotation=45, ha='right')
    ax3.set_ylabel('P(Feature)')
    ax3.set_title('P(Feature) - Marginal')
    ax3.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_conditional_probabilities_numeric(feature_col, prediction_col, feature_name):
    """
    Plot conditional probabilities for numeric features using KDE
    """
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'Conditional Probabilities: {feature_name}', fontsize=16, fontweight='bold')
    
    pred_vals = sorted(prediction_col.unique())
    colors = plt.cm.Set2(np.linspace(0, 1, len(pred_vals)))
    
    # Plot 1: P(feature | prediction) - KDE for each prediction class
    ax1 = axes[0]
    for i, pred in enumerate(pred_vals):
        mask = prediction_col == pred
        data = feature_col[mask].dropna()
        if len(data) > 1:
            data.plot.kde(ax=ax1, label=f'Grade {pred}', color=colors[i], linewidth=2)
    ax1.set_xlabel('Feature Value')
    ax1.set_ylabel('Density')
    ax1.set_title('P(Feature | Prediction)')
    ax1.legend()
    ax1.grid(alpha=0.3)
    
    # Plot 2: P(prediction | feature) - binned feature values
    ax2 = axes[2]
    # Create bins for the feature
    n_bins = min(10, len(feature_col.dropna().unique()))
    feature_bins = pd.qcut(feature_col.dropna(), q=n_bins, duplicates='drop')
    
    # For each bin, compute distribution of predictions
    bin_labels = feature_bins.cat.categories
    pred_distributions = np.zeros((len(bin_labels), len(pred_vals)))
    
    for i, bin_label in enumerate(bin_labels):
        mask = feature_bins == bin_label
        bin_predictions = prediction_col[feature_col[mask].index]
        dist = bin_predictions.value_counts(normalize=True)
        for j, pred in enumerate(pred_vals):
            pred_distributions[i, j] = dist.get(pred, 0)
    
    # Stacked bar chart
    bottom = np.zeros(len(bin_labels))
    for j, pred in enumerate(pred_vals):
        ax2.bar(range(len(bin_labels)), pred_distributions[:, j], 
                bottom=bottom, label=f'Grade {pred}', color=colors[j], alpha=0.8)
        bottom += pred_distributions[:, j]
    
    ax2.set_xticks(range(len(bin_labels)))
    ax2.set_xticklabels([f'{i+1}' for i in range(len(bin_labels))], rotation=45)
    ax2.set_xlabel('Feature Bin')
    ax2.set_ylabel('P(Prediction | Feature)')
    ax2.set_title('P(Prediction | Feature Bin)')
    ax2.legend()
    ax2.grid(axis='y', alpha=0.3)
    
    # Plot 3: P(feature) - marginal distribution
    ax3 = axes[1]
    feature_col.dropna().plot.kde(ax=ax3, color='navy', linewidth=2)
    ax3.fill_between(ax3.lines[0].get_xdata(), ax3.lines[0].get_ydata(), alpha=0.3)
    ax3.set_xlabel('Feature Value')
    ax3.set_ylabel('Density')
    ax3.set_title('P(Feature) - Marginal')
    ax3.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def analyze_all_features(df, prediction_column='damage_grade', exclude_cols=None):
    """
    Analyze all features in the dataframe
    """
    if exclude_cols is None:
        exclude_cols = ['building_id', prediction_column]
    
    prediction_col = df[prediction_column]
    
    for col_name in df.columns:
        if col_name in exclude_cols:
            continue
        
        print(f"\n{'='*60}")
        print(f"Analyzing: {col_name}")
        print(f"{'='*60}")
        
        feature_col = df[col_name]
        
        if pd.api.types.is_bool_dtype(feature_col.dtype):
            plot_conditional_probabilities_bool(feature_col, prediction_col, col_name)
        elif pd.api.types.is_categorical_dtype(feature_col.dtype) or pd.api.types.is_object_dtype(feature_col.dtype):
            plot_conditional_probabilities_categorical(feature_col, prediction_col, col_name)
        elif pd.api.types.is_numeric_dtype(feature_col.dtype):
            plot_conditional_probabilities_numeric(feature_col, prediction_col, col_name)

# Example usage:
if __name__ == "__main__":
    # Load and merge data
    df = load_and_merge_data('train_values.csv', 'train_labels.csv')
    
    # Perform analysis
    analyze_all_features(df, prediction_column='damage_grade')
```
